### **DATA PREPROCESSING**

```python
import os
import wfdb
import numpy as np
import h5py
import scipy.signal as sgn

def remove_baseline_filter(sample_rate):
    """For a given sampling rate"""
    fc = 0.8  # [Hz], cutoff frequency
    fst = 0.2  # [Hz], rejection band
    rp = 0.5  # [dB], ripple in passband
    rs = 40  # [dB], attenuation in rejection band
    wn = fc / (sample_rate / 2)
    wst = fst / (sample_rate / 2)

    filterorder, _ = sgn.ellipord(wn, wst, rp, rs)
    sos = sgn.iirfilter(filterorder, wn, rp, rs, btype="high", ftype="ellip", output="sos")
    return sos

def normalize(ecg, sample_rate):
    """Normalize ECG data."""
    sos = remove_baseline_filter(sample_rate)
    ecg_nobaseline = sgn.sosfiltfilt(sos, ecg, padtype="constant", axis=-1)

    new_freq = 400
    ecg_resampled = sgn.resample_poly(ecg_nobaseline, up=new_freq, down=sample_rate, axis=-1)

    new_len = 4096
    n_leads, length = ecg_resampled.shape
    ecg_zeropadded = np.zeros([n_leads, new_len])
    pad = (new_len - length) // 2
    ecg_zeropadded[..., pad: length + pad] = ecg_resampled

    return ecg_zeropadded

def load_ecg_data(dat_file, hea_file):
    """Load ECG data from .dat and .hea files."""
    record = wfdb.rdrecord(dat_file[:-4], sampto=4096)
    ecg_data = record.p_signal.T
    return ecg_data

def save_to_hdf5(ecg_data, age, sex, label, hdf5_file):
    """Save preprocessed ECG data, age, sex, and label to HDF5."""
    n_samples = ecg_data.shape[0]  # Number of ECG samples
    with h5py.File(hdf5_file, 'w') as f:
        f.create_dataset('x_ecg_test', data=ecg_data)  # Shape (n_samples, 12, 4096)
        f.create_dataset('x_age_test', data=np.full((n_samples,), age, dtype=np.int32))  # Shape (n_samples,)
        f.create_dataset('x_is_male_test', data=np.full((n_samples,), sex, dtype=np.int32))  # Shape (n_samples,)
        f.create_dataset('y_test', data=np.full((n_samples,), label, dtype=np.int32))  # Shape (n_samples,)

def preprocess_data(dat_file, hea_file, hdf5_file, age, sex, label):
    """Preprocess ECG data and save to HDF5."""
    ecg_data = load_ecg_data(dat_file, hea_file)
    ecg_data = normalize(ecg_data, sample_rate=500)
    save_to_hdf5(ecg_data, age, sex, label, hdf5_file)

# Example usage
if __name__ == "__main__":
    dat_file = "/content/nstemi_prediction/data/Test_data/1.dat"  # Replace with your .dat file path
    hea_file = "/content/nstemi_prediction/data/Test_data/1.hea"  # Replace with your .hea file path
    hdf5_file = "/content/nstemi_prediction/data/Processed_data/1.h5"  # HDF5 file path

    # Get age, sex, and label from user input
    age = int(input("Enter the age of the patient: "))
    sex = int(input("Enter the sex of the patient (0 for male, 1 for female): "))  # Check model convention
    label = int(input("Enter the label for the patient (0 for Control, 1 for STEMI, 2 for NSTEMI): "))  # Example labels

    preprocess_data(dat_file, hea_file, hdf5_file, age, sex, label)
```

### **MODEL PREDICTION**

```python
import os
import torch
import json
import pandas as pd
from torch.utils.data import DataLoader
from dataloader import MyECGDataset
from model import EnsembleECGModel
from tqdm import tqdm

import argparse
from argparse import Namespace
import torch.nn.functional as F
import numpy as np


def main(input_data, log_dir):
    # Read config file
    config_path = os.path.join(log_dir, "config.json")
    with open(config_path) as json_file:
        mydict = json.load(json_file)

    # Create a config object
    config = Namespace(**mydict)
    config.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # Load dataset
    dataset = MyECGDataset(input_data)
    test_loader = DataLoader(dataset, batch_size=config.batch_size, shuffle=False)

    # Load model
    model = EnsembleECGModel(config, log_dir)
    model.to(config.device)
    model.eval()

    # Prepare to collect results
    all_probs = []
    all_labels = []
    all_ids = []

    # Test run
    pbar = tqdm(test_loader, total=len(test_loader), desc="Test")
    for batch_idx, batch in enumerate(pbar):
        # Extract data from batch
        traces, labels, ids, age, sex = batch
        traces = traces.to(device=config.device)
        labels = labels.to(device=config.device)
        age_sex = torch.stack([sex, age]).t().to(device=config.device)

        # Forward pass
        with torch.no_grad():
            inp = (traces, age_sex)
            logits = model(inp)
            probs = torch.softmax(logits, dim=-1)

        all_probs.append(probs.detach().cpu())
        all_labels.append(labels.detach().cpu())
        all_ids.append(ids)

    # Concatenate results
    all_probs = torch.cat(all_probs, dim=0)
    all_labels = torch.cat(all_labels, dim=0)
    all_ids = torch.cat(all_ids, dim=0)

    # Save results to CSV
    df = pd.DataFrame({
        "ids": all_ids.numpy(),
        "labels": all_labels.numpy(),
        "logits": list(all_probs.numpy())
    })
    df.to_csv(os.path.join(log_dir, "logits.csv"), index=False)
    tqdm.write("Logits saved to csv file.")

if __name__ == "__main__":
    os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"

    # Specify input data and log directory
    input_data = "/content/nstemi_prediction/data/Processed_data/1.h5"  # HDF5 file path
    log_dir = "/content/nstemi_prediction/logs"  # Path to your model weights

    # Run the main function
    main(input_data, log_dir)
```
